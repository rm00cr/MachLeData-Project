```python
import kagglehub
import psycopg2
from sqlalchemy import create_engine

# psycopg2 is already installed in CELL INDEX 0, so no need to install it again
# Ensure psycopg2 is imported correctly
# No additional code is needed here as psycopg2 is already imported in this cell
```

```python
import pandas as pd
```

```python
def download_dataset():
    """
    Download the dataset from Kaggle.
    """
    dataset_path = kagglehub.dataset_download("samithsachidanandan/the-global-ai-ml-data-science-salary-for-2025")
    return dataset_path
 
def load_data(dataset_path):
    """
    Load the dataset from the specified path.
 
    Args:
        dataset_path (str): Path to the dataset.
 
    Returns:
        pd.DataFrame: Loaded dataset.
    """
    # Load the dataset
    data = pd.read_csv(dataset_path + "/salaries.csv")
    print(data.columns)
    return data
 
```

```python
dataset_path = download_dataset()
```

```python
df = load_data(dataset_path)
```

```python
connection = psycopg2.connect(
        host="localhost",
        database="mydatabase",
        user="myuser",
        password="mypassword",
        port="5434"
    )
```

```python
from sqlalchemy import text
import pandas as pd

# Execute the query to select all rows from the salaries table
query = "SELECT * FROM salaries"

# Use the connection to create a cursor and execute the query
cursor = connection.cursor()
cursor.execute(query)

# Fetch the results and column names
columns = [desc[0] for desc in cursor.description]
selected_data = pd.DataFrame(cursor.fetchall(), columns=columns)

# Print the first few rows of the DataFrame
print(selected_data.head())
```

```python


# Create a connection string for SQLAlchemy
connection_string = f"postgresql://{user}:{password}@{host}:{port}/{database}"

# Create an SQLAlchemy engine
engine = create_engine(connection_string)

# Insert data into the PostgreSQL table in batches
batch_size = 100  # Number of rows per batch
for i in range(0, len(df), batch_size):
    batch = df.iloc[i:i+batch_size]
    batch.to_sql('salaries', engine, if_exists='append', index=False)
    print(f"Inserted rows {i} to {i+len(batch)-1}")

print("Data has been inserted into the PostgreSQL table successfully!")
```
